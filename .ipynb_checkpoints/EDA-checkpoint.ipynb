```python
var = np.zeros(ser.shape)
mu = np.zeros(ser.shape)
```

```python
var[0] = (0.20/math.sqrt(252))**2
mu[0] = 0
```

```python
span = 20
lam = 2/(span+1)
for t in range(1,ser.shape[0]):
    mu[t] = lam*ser[t] + (1-lam)*mu[t-1]
    var[t] = lam*((ser[t]-mu[t])**2)+(1-lam)*var[t-1]
```

```python
time_horizons = [1,5,20,60,252]
```

```python
data = dict()
span=20
for h in time_horizons:
    ret = df.pct_change(h).dropna()
    sigma = ret.ewm(span=span, adjust=False).std(bias=True)
    data[h] = ret / (sigma * np.sqrt(h))
```

```python
df1 = pd.concat(data, axis=1).dropna()
```

```python
T_s, T_l = 8, 24
macd   = df1.ewm(span=T_s, adjust=False).mean() - df1.ewm(span=T_l, adjust=False).mean()
q      = macd / df1.rolling(63).std()
signal = q / q.rolling(252).std()
signal = signal.dropna()
```

```python
df2 = pd.concat({'ret': df1, 'macd': signal}, axis=1)
```

```python
tickers = df2.columns.get_level_values(2).unique()   # level-2 labels, in order
chosen  = tickers[rand_assets]                       # positions -> labels
df3 = df2.loc[:, (slice(None), time_horizons[1], list(chosen))].dropna()
```

```python
LOOKBACK = 20
P=15
S=3
```

```python
class ReturnDataset(Dataset):
  """
  For index i in [0, T - LOOKBACK - 1]:
    X = arr[i : i+LOOKBACK]    shape (LOOKBACK, 3)  — the history window
    y = arr[i + LOOKBACK]      shape (3,)            — the target next day
  """
  def __init__(self, arr):
      self.arr = torch.tensor(arr, dtype=torch.float32)

  def __len__(self):
      return len(self.arr) - LOOKBACK

  def __getitem__(self, i):
      return self.arr[i : i + LOOKBACK], self.arr[i + LOOKBACK]


# Chronological 70 / 15 / 15 split — never shuffle time series for splitting
vals   = df3.values
T      = len(vals)
n_tr   = int(0.70 * T)
n_val  = int(0.15 * T)

# Overlap the boundaries by LOOKBACK so each split has full windows
train_ds = ReturnDataset(vals[:n_tr])
val_ds   = ReturnDataset(vals[n_tr - LOOKBACK : n_tr + n_val])
test_ds  = ReturnDataset(vals[n_tr + n_val - LOOKBACK :])

BATCH    = 32
train_dl = DataLoader(train_ds, batch_size=BATCH, shuffle=True)
val_dl   = DataLoader(val_ds,   batch_size=BATCH, shuffle=False)
test_dl  = DataLoader(test_ds,  batch_size=BATCH, shuffle=False)

print(f"train={len(train_ds)}  val={len(val_ds)}  test={len(test_ds)}")
```

```python
class LPATCHTST(nn.Module):
  def __init__(self, input_size=3, hidden_size=20, num_layers=2, dropout=0.2):
      super().__init__()

      # eq 52: one independent LSTM per feature i = 1,...,d
      self.lstms = nn.ModuleList([
          nn.LSTM(1, hidden_size, num_layers, batch_first=False,
                  dropout=dropout if num_layers > 1 else 0.0)
          for _ in range(input_size)
      ])

      # each patch token = P steps × d features × hidden flattened
      # 15 × 3 × 20 = 900 → project down to d_model
      self.patch_proj = nn.Linear(P * input_size * hidden_size, hidden_size)

      self.encoder = nn.TransformerEncoder(
          nn.TransformerEncoderLayer(d_model=hidden_size, nhead=4, batch_first=False),
          num_layers=2
      )
      
      num_patches = (LOOKBACK - P) // S + 1   # 3
      self.head = nn.Linear(num_patches * hidden_size, input_size)  # Linear(60 → 3)

  def forward(self, x):
      # x: (seq_len, batch, d)
      B = x.shape[1]

      # --- eq 52: collect h_t^(i) at every time step for each feature ---
      all_hidden = []
      for i, lstm in enumerate(self.lstms):
          h = torch.zeros(lstm.num_layers, B, lstm.hidden_size, device=x.device)
          c = torch.zeros(lstm.num_layers, B, lstm.hidden_size, device=x.device)
          h_seq = []
          for t in range(x.shape[0]):
              _, (h, c) = lstm(x[t:t+1, :, i:i+1], (h, c))
              h_seq.append(h[-1])          # last layer's h_t: (batch, hidden)
          all_hidden.append(torch.stack(h_seq, dim=0))   # (seq_len, batch, hidden)

      # h_t^(i) is now explicit for every t and every feature i
      H = torch.stack(all_hidden, dim=2)
      # (seq_len=20, batch, d=3, hidden=20)

      # --- eq 53: segment into temporal patches and pass to Transformer ---
      patches = H.unfold(0, P, S)
      # (num_patches=3, batch, d=3, hidden=20, P=15)

      N, B_, D, Hid, _ = patches.shape
      patch_tokens = self.patch_proj(patches.flatten(2, 4))
      # (num_patches=3, batch, hidden=20)

      encoded = self.encoder(patch_tokens)
      # (num_patches=3, batch, hidden=20)

      flat = encoded.permute(1, 0, 2).flatten(1)
      # (batch, num_patches*hidden = 60)

      return self.head(flat)
```

```python
device    = torch.device("cuda" if torch.cuda.is_available() else "xpu")
model     = LPATCHTST(input_size=df3.shape[1]).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
criterion = nn.MSELoss()
EPOCHS    = 13

train_losses, val_losses = [], []

with tqdm(total=EPOCHS, desc="Training") as pbar:
    for epoch in range(1, EPOCHS + 1):
      # --- train ---
      model.train()
      total = 0.0
      for X, y in train_dl:
          # DataLoader gives (batch, seq, feat); LSTM wants (seq, batch, feat)
          X = X.permute(1, 0, 2).to(device)
          y = y.to(device)
          optimizer.zero_grad()
          loss = criterion(model(X), y)
          loss.backward()
          nn.utils.clip_grad_norm_(model.parameters(), 1.0)  # prevent exploding gradients
          optimizer.step()
          total += loss.item() * y.size(0)
      train_losses.append(total / len(train_ds))
    
      # --- validate ---
      model.eval()
      total = 0.0
      with torch.no_grad():
          for X, y in val_dl:
              X = X.permute(1, 0, 2).to(device)
              loss = criterion(model(X), y.to(device))
              total += loss.item() * y.size(0)
      val_losses.append(total / len(val_ds))
      pbar.set_postfix(loss=f"{val_losses[-1]:.4f}")  
      pbar.update(1)
    
      # if epoch % 10 == 0:
      #     print(f"epoch {epoch:3d}  train={train_losses[-1]:.4f}  val={val_losses[-1]:.4f}")
```

```python
# Loss curve
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(train_losses, label="train MSE")
ax.plot(val_losses,   label="val MSE")
ax.set_xlabel("epoch"); ax.set_ylabel("MSE"); ax.legend()
plt.tight_layout(); plt.show()

# Test predictions
model.eval()
preds, targets = [], []
with torch.no_grad():
  for X, y in test_dl:
      preds.append(model(X.permute(1, 0, 2).to(device)).cpu())
      targets.append(y)

preds   = torch.cat(preds).numpy()    # (N_test, 3)
targets = torch.cat(targets).numpy()  # (N_test, 3)

test_mse = ((preds - targets) ** 2).mean()
# Information coefficient — rank correlation between prediction and realized return
ic = pd.DataFrame(preds).corrwith(pd.DataFrame(targets)).mean()
print(f"Test MSE: {test_mse:.4f}  |  Mean IC: {ic:.4f}")
```
